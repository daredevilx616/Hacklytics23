```python
import pandas as pd
import matplotlib.pyplot as plt

import folium
from geopy.geocoders import Nominatim
```

```python
# Import the blood level concentration dataset
blc = pd.read_csv("./datasets/kaggle/lead-blood-usa-children.csv")
```

```python
# Print the data
print(blc)
```

```python
# Filter the data for the 95th percentile
blc_95th = blc[blc['Entity'] == '95th percentile']

# Filter the data for the median
blc_median = blc[blc['Entity'] == 'Median']

# Create a line plot of the 95th percentile blood lead concentration by year
plt.plot(blc_95th['Year'], blc_95th['Blood lead concentration (1-5 year olds)'], '-o', label='95th Percentile')

# Create a line plot of the median blood lead concentration by year
plt.plot(blc_median['Year'], blc_median['Blood lead concentration (1-5 year olds)'], '-o', label='Median')

# Add labels and title to the plot
plt.xlabel('Year')
plt.ylabel('Blood Lead Concentration')
plt.title('Blood Lead Concentration by Year')
plt.legend()

# Show the plot
plt.show()
```

```python
# Import the dataset
lpr = pd.read_csv("./datasets/kaggle/legal-controls-lead-paint.csv")
```

```python
print (lpr)
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")

def get_coordinates(country):
    location = geolocator.geocode(country)
    return (location.latitude, location.longitude)

lpr["coordinates"] = lpr["Entity"].apply(get_coordinates)
lpr[["Latitude", "Longitude"]] = pd.DataFrame(lpr["coordinates"].tolist(), index=lpr.index)
```

```python
import json

# Load the JSON data from the file
with open("./datasets/world-countries.json", "r") as file:
    data = json.load(file)

# Check the type of the data to confirm it's a Topology
if data["type"] == "Topology":
    print("The data is of type Topology")
else:
    print("The data is not of type Topology")

# Extract the arcs array from the data
arcs = data["arcs"]

# Print the number of arcs
print(f"Number of arcs: {len(arcs)}")

# Iterate over the arcs array and print each arc
for i, arc in enumerate(arcs):
    print(f"Arc {i + 1}: {arc}")

```

```python
# Create a folium map centered at the location
m = folium.Map(location=[45.5236, -122.6750], zoom_start=2)

# Iterate over the rows in the dataframe
for i, row in lpr.iterrows():
    # Get the country name and regulation status
    country = row["country"]
    regulated = row["regulated"]

    # Get the latitude and longitude of the country
    lat, lon = row["latitude"], row["longitude"]

    # Color the marker green or red depending on the regulation status
    if regulated:
        color = "green"
    else:
        color = "red"

    # Add a marker for the country on the map
    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.8,
        popup=country
    ).add_to(m)

# Show the map
m
```
